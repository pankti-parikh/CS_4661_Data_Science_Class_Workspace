# Question1: Predicting Heart Disease

## a- Read the dataset file “Hearts_s.csv” (you should download it from CSNS), and assign it to a Pandas DataFrame.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score

import numpy as np
import pandas as pd

heart_data = pd.read_csv('S:\MASTERS\SEMESTER_4\CS_4661\HomeWork\Homework_3\Heart_s.csv')
heart_data[0::10]
```

## b - Check out the dataset. As you see, the dataset contains a number of features including both contextual and biological factors(e.g. age, gender, vital signs, …). The last column“AHD” is the label with “Yes” meaning that a human subject has Heart  Disease, and “No” meaning that the subject does not have Heart Disease.

```python
def categorical_to_numerical(x):
    if x == 'Yes':
        return 1
    elif x == 'No':
        return 0
heart_data['label'] = heart_data['AHD'].apply(categorical_to_numerical)

y = heart_data['label']
y.head()
```

## c- As you see, there are at least 3 categorical features in the dataset (Gender, ChestPain,Thal). Let’s ignore these categorical features for now, only keep the numerical features and build your feature matrix and label vector.

```python
feature_cols = ['Age', 'RestBP', 'Chol', 'RestECG', 'MaxHR', 'Oldpeak']
    
X = heart_data[feature_cols] 
X.head()
```

## d- Split the dataset into testing and training sets with the following parameters: test_size=0.25, random_state=4.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=4)
```

## e- Use KNN (with k=3), Decision Tree (with random_state=5), and Logistic Regression Classifiers to predict Heart Disease based on the training/testing datasets that you built in part (d). Then check, compare, and report the accuracy of these 3 classifiers. Which one is the best? Which one is the worst?

```python
k=3
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(X_train, y_train)
y_predict_knn = knn.predict(X_test)
accuracy_score_knn = accuracy_score(y_test, y_predict_knn)
print('accuracy for KNN: ', accuracy_score_knn )

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_predict_lr = logreg.predict(X_test)
accuracy_score_lr = accuracy_score(y_test, y_predict_lr)
print('accuracy for Logistic Regression: ', accuracy_score_lr)

decisiontree = DecisionTreeClassifier(random_state=5)
decisiontree.fit(X_train, y_train)
y_predict_dt = decisiontree.predict(X_test)
accuracy_score_dt = accuracy_score(y_test, y_predict_dt)
print('accuracy for Decisiontree: ', accuracy_score_dt)
```

### BEST CLASSIFIER IS  : Logistic Regression with accuracy = 0.7236842105263158

### WORST CLSSIFIER IS  : Decisiontree  with accuracy = 0.6052631578947368

## f-Now, we want to use the categorical features as well! To this end, we have to perform a feature engineering process called OneHotEncoding for the categorical features. To do this, each categorical feature should be replaced with dummy columns in the feature table (one column for each possible value of a categorical feature), and then encode it in a binary manner such that only one of the dummy columns can take “1” at a time (and zero for the rest). For example, “Gender” can take two values “m” and “f”. Thus, we need to replace this feature (in the feature table) by 2 columns titled “m” and “f”. Wherever we have a male subject, we can put “1” and ”0” in the columns “m” and “f”. Wherever we have a female subject, we can put “0” and ”1” in the columns “m” and “f”. (Hint: you will need 4 columns to encode “ChestPain” and 3 columns to encode “Thal”).

```python
# OneHotEncoding 
heart_data['male'] = heart_data['Gender'].map(lambda x: 1 if x=='m' else 0)
heart_data['female'] = heart_data['Gender'].map(lambda x: 1 if x=='f' else 0)

heart_data['typical'] = heart_data['ChestPain'].map(lambda x: 1 if x=='typical' else 0)
heart_data['asymptomatic'] = heart_data['ChestPain'].map(lambda x: 1 if x=='asymptomatic' else 0)
heart_data['nontypical'] = heart_data['ChestPain'].map(lambda x: 1 if x=='nontypical' else 0)
heart_data['nonanginal'] = heart_data['ChestPain'].map(lambda x: 1 if x=='nonanginal' else 0)

heart_data['fixed'] = heart_data['Thal'].map(lambda x: 1 if x=='fixed' else 0)
heart_data['normal'] = heart_data['Thal'].map(lambda x: 1 if x=='normal' else 0)
heart_data['reversable'] = heart_data['Thal'].map(lambda x: 1 if x=='reversable' else 0)

heart_data[0::10]
```

## g-Repeat parts (d) and (e) with the new dataset that you built in part (f). How does the prediction accuracy change for each method?

```python
new_feature_cols = ['Age','RestBP','Chol','RestECG','MaxHR','Oldpeak', 'male', 'female', 'typical', 'asymptomatic',
                'nontypical', 'nonanginal','fixed', 'normal', 'reversable']

X1 = heart_data[new_feature_cols]
X_train, X_test,y_train,y_test = train_test_split(X1, y, test_size=0.25, random_state=4)

k=3
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(X_train, y_train)
y_predict_knn = knn.predict(X_test)
score_knn = accuracy_score(y_test, y_predict_knn)
print('accuracy for KNN:', score_knn)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_predict_lr = logreg.predict(X_test)
score_lr = accuracy_score(y_test, y_predict_lr)
print('accuracy for Logistic Regression:', score_lr)

decisiontree = DecisionTreeClassifier(random_state=5)
decisiontree.fit(X_train, y_train)
y_predict_dt = decisiontree.predict(X_test)
score_dt = accuracy_score(y_test, y_predict_dt)
print('accuracy for Decisiontree: ', score_dt)
```

## h- Now, repeat part (e) with the new dataset that you built in part (f), but this time using Cross-Validation. Thus, rather than splitting the dataset into testing and training, use 10-fold Cross-Validation (as we learned in Lab4) to evaluate the classification methods and report the final prediction accuracy.

```python
from sklearn.cross_validation import cross_val_score

accuracy_list = cross_val_score(knn, X1, y, cv=10, scoring='accuracy')
print(accuracy_list)
accuracy_cv = accuracy_list.mean()
print('\n')
print('accuracy for Logistic KNN: ', accuracy_cv)

print("-------------------------------------------------------------------------------------------------------------")
accuracy_list = cross_val_score(logreg, X1, y, cv=10, scoring='accuracy')
print(accuracy_list)
accuracy_cv = accuracy_list.mean()
print('\n')
print('accuracy for Logistic Regression: ', accuracy_cv)

print("-------------------------------------------------------------------------------------------------------------")
accuracy_list = cross_val_score(decisiontree, X1, y, cv=10, scoring='accuracy')
print(accuracy_list)
accuracy_cv = accuracy_list.mean()
print('\n')
print('accuracy for Decisiontree: ', accuracy_cv)
```
